```python
boards_default = (21,
 4,
 5,
 17,
 2,
 3,
 25,
 9,
 10,
 15,
 22,
 7,
 11,
 19,
 12,
 24,
 16,
 23,
 20,
 1,
 6,
 13,
 8,
 14,
 18)
```

```python
boards = list(boards_default)
def generate_bingo_board(num, _boards):
    index = 0
    for i in _boards:
        if _boards[index] == num:
            _boards[index] = 'XX'
        index += 1
    return _boards

def print_bingo_board(_boards):
    index = 1
    end = ''
    for i in _boards:
        if index % 5 == 0:
            end = '\n'
        else:
            end = ''
        print(f"{str(i).zfill(2)} ", end=end)
        index += 1
    print('________________________________')
    
#用array裝位置
def get_bingo_loctions(_boards):
    index = 1 
    ans = []
    for i in _boards:
        if i == 'XX':
            ans.append(index)
        index += 1
    print(ans)
    
    return ans
    
def get_bingo_count(_guess, _ans):
    ans_point = 0
    for i in _ans:
        if i in _guess:
            ans_point += 1

    if ans_point == len(_ans):
        return 1
    return 0

a1 =[1,2,3,4,5]
a2= [6,7,8,9,10]
a3=[11,12,13,14,15]
a4=[16,17,18,19,20]
a5=[21,22,23,24,25]
b1=[1,6,11,16,21]
b2=[2,7,12,17,22]
b3=[3,8,13,18,23]
b4=[4,9,14,19,24]
b5=[5,10,15,20,25]
c1=[1,7,13,19,25]
c2=[5,9,13,17,21]
total_ans = [a1,a2,a3,a4,a5,b1,b2,b3,b4,b5,c1,c2]
```

```python
target_bingo_number = 3
bingo = 0
bypass_ans= []
while bingo != target_bingo_number:
    choice = input('輸入下一個數字:')
    
    bingo_choice = generate_bingo_board(int(choice), boards)
    print_bingo_board(boards)
    my_ans = get_bingo_loctions(boards) 
    print(my_ans)
    for ans in total_ans:
        print('before:',bingo,end='')
        print(ans)
        if ans not in bypass_ans:
            bingo_count = get_bingo_count(my_ans, ans)
            if bingo_count:
                bingo += bingo_count
                #將此組賓果從 total_ans 移除
                bypass_ans.append(ans)
            
        print('after: ',bingo)
        if bingo == target_bingo_number:
            print('bingo!!')
            break
```
